```python
df['variable'].unique()
```

```python
df_number = df[df['variable'] == 'Excess deaths (number)']
df_percent = df[df['variable'] == 'Excess deaths (% change from average)']
```

```python
import matplotlib.pyplot as plt

country_totals = df_number.groupby('country')['value'].sum().sort_values(ascending=False).head(10)

country_totals.plot(kind='bar')
plt.title("Top 10 Countries by Total Excess Deaths")
plt.ylabel("Total Excess Deaths")
plt.xticks(rotation=90)
plt.show()
```

```python
yearly_trend = df_number.groupby('year')['value'].sum()

yearly_trend.plot(marker='o')
plt.title("Total Excess Deaths by Year")
plt.ylabel("Excess Deaths")
plt.xlabel("Year")
plt.show()
```

```python
weekly_trend = df_number.groupby('week_number')['value'].sum()

weekly_trend.plot()
plt.title("Weekly Excess Deaths Pattern")
plt.xlabel("Week Number")
plt.ylabel("Excess Deaths")
plt.show()
```

```python
gender_totals = df_number.groupby('gender')['value'].sum()

gender_totals.plot(kind='bar')
plt.title("Total Excess Deaths by Gender")
plt.ylabel("Excess Deaths")
plt.show()
```

```python
age_totals = df_number.groupby('age')['value'].sum().sort_values(ascending=False)

age_totals.plot(kind='bar', figsize=(10,5))
plt.title("Excess Deaths by Age Group")
plt.ylabel("Excess Deaths")
plt.xticks(rotation=90)
plt.show()
```

```python
import seaborn as sns

pivot = df_number.pivot_table(values='value',
                              index='week_number',
                              columns='year',
                              aggfunc='sum')

plt.figure(figsize=(10,6))
sns.heatmap(pivot)
plt.title("Weekly Excess Deaths by Year")
plt.show()
```

```python
country_percent = df_percent.groupby('country')['value'].mean().sort_values(ascending=False).head(10)

country_percent.plot(kind='bar')
plt.title("Top 10 Countries by Average % Excess Deaths")
plt.ylabel("% Change from Average")
plt.show()
```
